```python
# constant variables
DATASET_FILE_NAME = "tech_collab_hub_dataset.xlsx"
```

```python
import requests
from bs4 import BeautifulSoup
import re
def getImageLink(repo_name):
    url1 = f"https://www.freepik.com/api/regular/search?filters[content_type]=vector&locale=en&term=${repo_name}"
    headers = {
    "Connection" : "keep-alive",
    "Accept" : "*/*",
    "Cookie":"csrf_freepik=2cc754229a066f7a03c17adbfe04f577",
    "User-Agent":"PostmanRuntime/7.37.0"
    }
    response = requests.get(url1, headers=headers)
    if response.status_code == 200:
        responsejson = response.json()
        response2 = requests.get(responsejson['items'][0]['url'], headers=headers)
        pattern = rf'https://img.freepik.com/free-vector/{responsejson['items'][0]['slug']}.*?\.jpg'
        match = re.search(pattern, response2.text)
        if match :
            return str(match.group())
        else:
            print('no match')
            return -1

    else : 
        print('error - ' + str(response.status_code))
        return response.status_code
    
```

```python
import pandas as pd
cnt = 1
df_for_iter = pd.read_excel(DATASET_FILE_NAME)
df = pd.read_excel(DATASET_FILE_NAME)
for index, row in df_for_iter.iterrows():
  if(row['no_of_contributors'] > 0):
    link = getImageLink(row['repo_name'])
    if isinstance(link, str):
      df.loc[index, 'project_icon'] = link
      print("done " + str(cnt), end=" ")
      print(link)
    else:
      print('not done ' + str(cnt) + " o/p=" + str(link))
  else :
    print("skip " + str(cnt))
  cnt+=1
df.to_excel(DATASET_FILE_NAME, index=False)
```
