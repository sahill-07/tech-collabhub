```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def addTopicLang(repo_link):
    print(repo_link)
    response = requests.get(repo_link)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        a_tag = soup.findAll('a', class_='topic-tag-link', attrs={'data-ga-click': 'Topic, repository page'})
        topic = []
        if a_tag :
            for item in a_tag:
                value = item.text.strip()
                topic.append(value)
        topic = ', '.join(topic)
        return topic
    elif response.status_code == 404:
      return 2

    else:
        print(f"Failed to retrieve website page at {repo_link}. Status code: {response.status_code}")
        return 0
    # break
```

```python
# Read the Excel file
import time
cnt = 1
df_for_iter = pd.read_excel('tech_collab_hub_dataset.xlsx')
df = pd.read_excel('tech_collab_hub_dataset.xlsx')
# Iterate through each row of the DataFrame
for index, row in df_for_iter.iterrows():
  response = addTopicLang(row['repo_link'])
  if isinstance(response, str):
    df.loc[index, 'topic'] = response
    print("completed : " + str(cnt) + " data=" + response)
  elif response != 2:
    while response == 0:
      time.sleep(10)
      response = addTopicLang(index, row)
    # print('completed : ' + str(cnt) + " data=" + response)
  else :
    print("skipped : " + str(cnt))
  time.sleep(1)
  cnt+=1
df.to_excel('tech_collab_hub_dataset.xlsx', index=False)
```
